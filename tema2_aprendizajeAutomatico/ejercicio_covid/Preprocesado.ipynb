

Para elminiar las filas en las que falta algún valor: `dropna`

*   Con `inplace=True` se modifica el dataframe, y no devuelve nada

    `df.dropna(inplace = True)`

*   Por defecto `inplace=False` y devuelve el dataframe con las filas eliminadas, que habrá que guardar en otro:

    `df2 = df2=df.dropna()`

```python
df2 = df2=df.dropna()
df2
```


Otra opción era rellenar los valores que faltan con un valor, por ejemplo 0, la media u otro, usando `fillna()`

```python
media = df["Calorías"].mean()
df["Calorías"].fillna(media, inplace=True)
df
```

Para arreglar los valores de fecha incorrectos, se puede intentar corregir con `to_datetime()`

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], errors = 'coerce')
df.dropna(subset=['Fecha'], inplace = True)
df
```

Si algún valor está fuera de un rango establecido se puede corregir.

Por ejemplo si el mayor valor de la duración es 120 los que sean mayores se ponen a este valor máximo:

```python
duracion_max = 120
for i in df.index:
    if df.loc[i, "Duración"] > duracion_max:
        df.loc[i, "Duración"] = duracion_max
df
```

También se pueden elmininar las filas duplicadas:

```python
print(df.duplicated())

df.drop_duplicates(inplace = True)
df
```
