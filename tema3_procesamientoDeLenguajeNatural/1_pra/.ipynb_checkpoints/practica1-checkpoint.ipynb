```python
import spacy

nlp = spacy.load('en_core_web_sm')
nlp.pipe_names

nlpEs = spacy.load('es_core_news_sm')
nlpEs.pipe_names
```

```python
#Imprimimos el tipo de cada uno de los token que componen la frase
doc = nlp("He comido una amburguesa grasienta.")
for token in doc:
    print(token.text, "-->", token.pos_)
```

```python
#Imprimimos las dependencias de cada uno de los token que componen la frase
for token in doc:
    print(token.text, "-->", token.dep_)

#Imprimimos el significado de cada una de las etiquetas
spacy.explain("nsubj"), spacy.explain("ROOT"), spacy.explain("aux"), spacy.explain("advcl"), spacy.explain("dobj"), 
```

```python
#Imprimimos las dependencias de forma gr√°fica
from spacy import displacy
displacy.render(doc, style='dep', jupyter=True)
```

```python
doc2 = nlp("He went to play basketball. He ate a burger. She drank a coke.")
displacy.render(doc2, style='dep', jupyter=True)

for token in doc2:
    print(token.text, "-->", token.dep_)
    print(token.lemma_)
    if token.n_rights>0 and token.n_lefts>0:
        print(" ", "CHRILDREN:")
        
        print(" ", "(left):")
        for subtoken in token.lefts:
            print("    ", subtoken)
            
        print(" ", "(left):")
        for subtoken in token.rights:
            print("    ", subtoken)
    print()
```
